**Quest Giver**: Head of Research and Development

**Quest Level**: Intermediate

**Quest Description**: Our lab has been hit by an unexpected viral outbreak. We need to clean up the entire facility before anyone can return safely. 1. Contact Vendor (Difficulty: Easy)
Visit the local industrial supplier and negotiate the purchase or lease of a top-of-the-line HEPA filtration unit suitable for large facilities like ours. Make sure you acquire any necessary permits for transport and installation if required.

2. Transport Filters (Difficulty: Moderate)
Arrange transportation for the massive filters to be delivered to our lab. You may have to engage specialized equipment due to their size and weight. Ensure proper safety protocols are followed during transit.

3. Install Base Framework (Difficulty: Challenging)
Assemble the base framework of the HEPA filtration system within the designated area at the lab. Follow instructions meticulously as incorrect assembly might result in malfunctioning later on.

4. Connect Filter Components (Difficulty: Hard)
Connect each individual component of the HEPA filter securely, ensuring they fit correctly into the frame. Double check every connection point to avoid leaks or breaches which could compromise the efficiency of the device.

5. Power Up & Test Run (Difficulty: Advanced)
Once everything is connected, power up the system and run diagnostics tests provided by the manufacturer. If there are no issues detected, proceed to the final step.

6. Full System Integration (Difficulty: Expert)
Integrate the new HEPA filtration system fully into our existing ventilation infrastructure. This process involves complex wiring work so please exercise extreme caution not to damage either system while connecting them together. After completion, perform another round of comprehensive testing to confirm successful integration.

**Rewards**: Acknowledgement letter from the R&D department head along with a bonus paycheck; permanent access to high-level biohazard suits used only for handling such crises.
